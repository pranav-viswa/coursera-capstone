<p>SCRAPPING THE WIKIPEDIA PAGE

```python
import requests # for http requests
import pandas as pd # for creating the dataframe
from bs4 import BeautifulSoup # for scrapping webdata
```

```python
page = requests.get("https://en.wikipedia.org/wiki/List_of_postal_codes_of_Canada:_M")
soup = BeautifulSoup(page.content, 'html.parser')
```

```python
table = soup.find('tbody')
rows = table.select('tr')
row = [r.get_text() for r in rows]
```

<p><b>Now let us clean the scrapped data to get a clean and untangled table .
 <p>table is then converted to a dataframe

```python
df = pd.DataFrame(row)
df1 = df[0].str.split('\n', expand=True)
df2 = df1.rename(columns=df1.iloc[0])
df3 = df2.drop(df2.index[0])
df3
```

<p><b>Inoring the cella that has NOT ASSIGNED IN BOROUGH COLUMN.

```python
df4 = df3[df3.Borough != 'Not assigned']
df4
```

<p><b>combining the neighbourhoods having the same postcode

```python
df5 = df4.groupby(['Postcode', 'Borough'], sort = False).agg(','.join)
df5.reset_index(inplace = True)
df5
```

```python
df6=df5.replace('Not assigned','queens park')
df6
```
